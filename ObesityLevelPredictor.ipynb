```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv("data.csv")
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates().sum()
```

```python
Gender_encoder=LabelEncoder()
CALC_encoder=LabelEncoder()
FAVC_encoder=LabelEncoder()
SCC_encoder=LabelEncoder()
SMOKE_encoder=LabelEncoder()
family_history_with_overweight_encoder=LabelEncoder()
CAEC_encoder=LabelEncoder()
MTRANS_encoder=LabelEncoder()
NObeyesdad_encoder=LabelEncoder()
data["Gender"]=Gender_encoder.fit_transform(data["Gender"])
data["CALC"]=CALC_encoder.fit_transform(data["CALC"])
data["FAVC"]=FAVC_encoder.fit_transform(data["FAVC"])
data["SCC"]=SCC_encoder.fit_transform(data["SCC"])
data["SMOKE"]=SMOKE_encoder.fit_transform(data["SMOKE"])
data["family_history_with_overweight"]=family_history_with_overweight_encoder.fit_transform(data["family_history_with_overweight"])
data["CAEC"]=CAEC_encoder.fit_transform(data["CAEC"])
data["MTRANS"]=MTRANS_encoder.fit_transform(data["MTRANS"])
data["NObeyesdad"]=NObeyesdad_encoder.fit_transform(data["NObeyesdad"])
data.head()
```

```python
x=data.drop("NObeyesdad",axis=1)
y=data[["NObeyesdad"]]
```

```python
x_scaler=StandardScaler()
x_scaled=x_scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
model=Sequential([Dense(16,input_shape=(16,)),
Dense(8,activation='relu'),
Dense(4,activation='relu'),
Dense(1,activation='sigmoid')])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=500)
```

```python
y_predict=model.predict(x_test)
accuracy=accuracy_score(y_test,y_predict)
print("accuracy =",accuracy*100,"%")
error=mean_squared_error(y_test,y_predict)
print("Error=",error)
print("RMS=",np.sqrt(error))
```

```python
joblib.dump(model,"obesity_trnd.pkl")
joblib.dump(x_scaler,"obesityXscaler.pkl")
joblib.dump(Gender_encoder,"Gender_encoder.pkl")
joblib.dump(CALC_encoder,"CALC_encoder.pkl")
joblib.dump(FAVC_encoder,"FAVC_encoder.pkl")
joblib.dump(SCC_encoder,"SCC_encoder.pkl")
joblib.dump(SMOKE_encoder,"SMOKE_encoder.pkl")
joblib.dump(family_history_with_overweight_encoder,"family_history_with_overweight_encoder.pkl")
joblib.dump(CAEC_encoder,"CAEC_encoder.pkl")
joblib.dump(MTRANS_encoder,"MTRANS_encoder.pkl")
joblib.dump(NObeyesdad_encoder,"NObeyesdad_encoder.pkl")
```
